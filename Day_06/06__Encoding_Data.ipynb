# Feature Encoding

Feature encoding is the process of transforming `categorical features` into `numeric features`. This is necessary because machine learning algorithms can only `handle numeric features`. There are many different ways to encode categorical features, and each method has its own advantages and disadvantages. 

In this notebook, we will explore some of the most popular methods for encoding categorical features, such as:

- ` Label encoding`
- `Ordinal encoding`
- ` One-hot encoding`
- `Binary encoding`

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
# data load
df = sns.load_dataset('tips')
df.head()
```

# __`01 Label-Encoding Method`__

```python
# Check the Values Count of Time coloumn
df['time'].value_counts()
```

```python
# let's encode the time in label-encoder with sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['encoded_time'] = le.fit_transform(df['time'])
df.head()
```

```python
# Check again the Values count now After Encoding
df['encoded_time'].value_counts()
```

#  __`02 Ordinal Encoding Method `__

```python
# Let's Apply For day Coloumn Before We Check The Value counts
df['day'].value_counts()
```

```python
# ordinal encoding the day column using specific order
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

# __`03 Onehot-Encoding Methodm`__

```python
df['sex'].value_counts()
```

```python
# one hot encoding on day column
ohe = OneHotEncoder()
ohe.fit_transform(df[['sex']]).toarray()
```

```python
# example of one hot encoding
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python

onehot_encoder = OneHotEncoder(sparse=False)
embarked_onehot = onehot_encoder.fit_transform(titanic[['embarked']])
```

```python
embarked_onehot_df = pd.DataFrame(embarked_onehot, columns=onehot_encoder.get_feature_names_out(['embarked']))
titanic = pd.concat([titanic.reset_index(drop=True), embarked_onehot_df.reset_index(drop=True)], axis=1)
titanic.head()
```
