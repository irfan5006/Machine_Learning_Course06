# **Inverse encoding / Inverse transform / Decoding categorical variables**

Inverse encoding is the process of converting the encoded data back to its original form. This is done by using the `inverse_transform()` method. 

This method is available for the following encoders:

- `LabelEncoder`
- `OrdinalEncoder`
- `OneHotEncoder`

```python
# import libraries
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

# **LabelEncoder**

A label encoder is a technique used in machine learning to convert `categorical data into numerical form`.\
In many machine learning algorithms, input data needs to be in numerical format, and label encoding is a `simple way` to achieve this for categorical variables.

```python
# load the data
df = sns.load_dataset('titanic')
df.head()
```

```python
# impute missing values for age, embarked, embark_town and deck
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)

# drop deck column
df.drop('deck', axis=1, inplace=True)
```

```python
# encode all categorical variables and object variables

le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_class = LabelEncoder()
le_who = LabelEncoder()
le_embark_town = LabelEncoder()

# Fit the Each Label Encoder of each column
df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['class'] = le_class.fit_transform(df['class'])
df['who'] = le_who.fit_transform(df['who'])
df['embark_town'] = le_embark_town.fit_transform(df['embark_town'])
df.head()
```

```python
# inverse transform the data

df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['class'] = le_class.inverse_transform(df['class'])
df['who'] = le_who.inverse_transform(df['who'])
df['embark_town'] = le_embark_town.inverse_transform(df['embark_town'])
df.head()
```

# **One Hot Encoding**

One-hot encoding is another technique used to represent `categorical data in a numerical format`.\
It is particularly useful when dealing with categorical variables that have no inherent ordinal relationship.\
The basic idea behind one-hot encoding is to represent each category with a `binary (0 or 1)` vector.


```python
import warnings
warnings.filterwarnings("ignore")

# Import Libraries
from sklearn.preprocessing import OneHotEncoder

# Load Dataset
df = sns.load_dataset('titanic')

# Categorical Columns
cat_columns = ['sex', 'embarked']

# One Hot Encoding
encoder = OneHotEncoder(sparse=False)
encoded_df = pd.DataFrame(encoder.fit_transform(df[cat_columns]))

# concatenate the dataframes 
df = pd.concat([df, encoded_df], axis=1)
# df.drop(cat_columns, axis=1, inplace=True)
df.head()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows of the datase
titanic.head()

# Extract the 'class' column
class_column = titanic[['class']]

# Create an OrdinalEncoder
class_ode = OrdinalEncoder(categories=[['First', 'Second', 'Third']])

# Fit and transform the 'class' column
titanic['class_encoded'] = class_ode.fit_transform(class_column)

# Display the DataFrame with the original 'class' and the encoded 'class'
print(titanic[['class', 'class_encoded']])



```

```python
# Reshape the 'class_encoded' column to have one dim# Reshape the 'class_encoded' column to have one dimension
titanic['class_decoded'] = class_ode.inverse_transform(titanic['class_encoded'].values.reshape(-1, 1)).flatten()

# Display the DataFrame with the original 'class', the encoded 'class', and the decoded 'class'
print(titanic[['class', 'class_encoded', 'class_decoded']])
```
